```python
def knn(X_train, X_test, y_train, y_test,n_neighbors=1,graph=False):
    knn_classifier = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    begin_time=datetime.now()
    knn_classifier.fit(X_train, y_train)
    end_time=datetime.now()
#     y_pred = knn_classifier.predict(X_test)
    train_score = knn_classifier.score(X_train, y_train)
    test_score = knn_classifier.score(X_test, y_test)
#     print(knn_classifier.predict(X_test))
#     print()
    return train_score,test_score,knn_classifier,end_time-begin_time
```

```python
from sklearn.svm import SVR
```

```python
def svm(X_train, X_test, y_train, y_test,kernal='rbf',graph=False):
    svm_regressor = SVR(kernel=kernal)
    begin_time=datetime.now()
    svm_regressor.fit(X_train, y_train)
    end_time=datetime.now()
#     y_pred_linear = svm_classifier.predict(X_test)
    train_score = svm_regressor.score(X_train, y_train)
    test_score = svm_regressor.score(X_test, y_test)
    return train_score,test_score,svm_regressor,end_time-begin_time
```

```python
def decision_tree(X_train, X_test, y_train, y_test,graph=False,graph_name="1"):
#     X_train, X_test, y_train, y_test = train_test_split(data_X, data_Y, test_size=0.3)
    clf = DecisionTreeClassifier()
    clf=clf.fit(X_train,y_train)
    train_times=[]
#     test_times=[]
    pruning_path=clf.cost_complexity_pruning_path(X_train,y_train)
#     print("ccp_alphas",pruning_path['ccp_alphas'])
#     print("impurities",pruning_path['impurities'])
    alphas=pruning_path['ccp_alphas']
    impurities=pruning_path['impurities']
    alphas_unique=list(set(alphas))
    alphas_unique.sort()
#     alphas_unique=alphas_unique[:-1]
    clfs=[]
    for alpha in alphas_unique:
        
        clf=DecisionTreeClassifier(ccp_alpha=alpha)
        begin_time=datetime.now()
        clf.fit(X_train,y_train)        
        end_time=datetime.now()
        train_times.append(end_time-begin_time)
        clfs.append(clf)
    train_scores = [clf.score(X_train, y_train) for clf in clfs]
    test_scores = [clf.score(X_test, y_test) for clf in clfs]
    # avoid system issue, only graph in my own and provide in report
    if graph:
        fig, ax = plt.subplots()
        ax.set_xlabel("Alpha")
        ax.set_ylabel("Accuracy")
        ax.set_title("Accuracy vs alpha for training and testing sets")
        ax.plot(alphas_unique, train_scores, marker="o", label="Train")
        ax.plot(alphas_unique, test_scores, marker="o", label="Test")
#         print(alphas_unique)
#         print(train_scores)
#         print(test_scores)
        ax.legend()
        # plt.show()
        plt.savefig("dt-"+graph_name+".png")
    return alphas_unique,train_scores,test_scores,clfs,train_times
# def test_decision_tree(data_X,data_Y)

def boosted_decision_tree(X_train, X_test, y_train, y_test,alpha=0.0,graph=False):
    gbc = GradientBoostingClassifier(ccp_alpha=alpha)
    begin_time=datetime.now()
    gbc.fit(X_train, y_train)     
    end_time=datetime.now()
#     y_pred = gbc.predict(X_test)
    train_score = gbc.score(X_train, y_train)
    test_score = gbc.score(X_test, y_test)
    return train_score,test_score,gbc,end_time-begin_time

def neural_network(X_train, X_test, y_train, y_test,graph=False,max_iter=700,size=70):
    mlp_clf = MLPClassifier(hidden_layer_sizes=(size, size), max_iter=max_iter)
    begin_time=datetime.now()
    mlp_clf.fit(X_train, y_train)
    end_time=datetime.now()
    # y_pred = mlp_clf.predict(X_test)
    train_score = mlp_clf.score(X_train, y_train)
    test_score = mlp_clf.score(X_test, y_test)
    return train_score,test_score,mlp_clf,end_time-begin_time
```

```python
def run_models(X_train, X_test, y_train, y_test,name="NowSecure"):
    models={}
    alphas_unique,train_scores_dt,test_scores_dt,clfs,train_times=decision_tree(X_train, X_test, y_train, y_test,graph=True,graph_name=name)
    
    print('- - - Decision Tree - - -')
    max_test_index=0
    max_test_score=0
    for i in range(len(alphas_unique)):
#         print(test_scores_dt[i])
#         print('Alpha : ',alphas_unique[i])
#         print('Train_score : ',train_scores_dt[i],'\tTest_score : ',test_scores_dt[i])
        if test_scores_dt[i]>=max_test_score:
            max_test_index=i
            max_test_score=test_scores_dt[i]
    best_dt=clfs[max_test_index]
    begin_time=datetime.now()    
    pred_y=best_dt.predict(X_test)
    end_time=datetime.now()
    print('Best Alpha : ',alphas_unique[max_test_index])
    print("Time cost - Train:",str(train_times[max_test_index]),"\tTest :",str(end_time-begin_time))
    print('Score - Train : ',train_scores_dt[max_test_index],'\tTest : ',test_scores_dt[max_test_index])   
    models['DT']=best_dt
    
    print('- - - Boosted Decision Tree with alpha '+str(alphas_unique[max_test_index])+' - - -')
    train_score_bdt,test_score_bdt,gbc,train_time = boosted_decision_tree(X_train, X_test, y_train, y_test,alpha=alphas_unique[max_test_index],graph=False)
    begin_time=datetime.now()    
    pred_y=gbc.predict(X_test)
    end_time=datetime.now()    
    print("Time cost - Train  :",str(train_time),"\tTest :",str(end_time-begin_time))
    print('Score - Train : ',train_score_bdt,'\tTest : ',test_score_bdt)
    models['BT']=test_score_bdt
    
    print('- - - Neural Network - - -')
    train_score_nn,test_score_nn,mlp_clf,train_time = neural_network(X_train, X_test, y_train, y_test,graph=False)
    begin_time=datetime.now()    
    pred_y=mlp_clf.predict(X_test)
    end_time=datetime.now()
    print("Time cost - Train  :",str(train_time),"\tTest :",str(end_time-begin_time))
    print('Score - Train : ',train_score_nn,'\tTest : ',test_score_nn)
    models['NN']=mlp_clf
        
    print('- - - Support Vector Machine (Linear Kernal) - - -')
    train_score_l,test_score_l,svm_classifier_l,train_time = svm(X_train, X_test, y_train, y_test,kernal='linear',graph=False)
    begin_time=datetime.now()    
    pred_y=svm_classifier_l.predict(X_test)
    end_time=datetime.now()    
    print("Time cost - Train  :",str(train_time),"\tTest :",str(end_time-begin_time))
    print('Score - Train : ',train_score_l,'\tTest : ',test_score_l)
    models['SVML']=svm_classifier_l
        
    print('- - - Support Vector Machine (RBF Kernal) - - -')
    train_score_rbf,test_score_rbf,svm_classifier_rbf,train_time = svm(X_train, X_test, y_train, y_test,kernal='rbf',graph=False)
    begin_time=datetime.now()    
    pred_y=svm_classifier_rbf.predict(X_test)
    end_time=datetime.now()
    print("Time cost - Train  :",str(train_time),"\tTest :",str(end_time-begin_time))
    print('Score - Train : ',train_score_rbf,'\tTest : ',test_score_rbf)
    models['SVMR']=svm_classifier_rbf
    return models
```
